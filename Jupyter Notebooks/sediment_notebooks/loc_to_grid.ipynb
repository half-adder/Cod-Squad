```python
import pandas as pd
```

```python
loc_sed = 'C:\\Users\\mdulieu\\Documents\\hackforthesea\\Cod-Squad\\data\\sediment\\sediment_loc.csv'
loc_grid = 'C:\\Users\\mdulieu\\Documents\\hackforthesea\\Cod-Squad\\data\\grid\\grids.csv'
dest_sed_grid = 'C:\\Users\\mdulieu\\Documents\\hackforthesea\\Cod-Squad\\data\\sediment\\sediment_loc_grid.csv'
```

```python
sediments = pd.read_csv(loc_sed)
grids = pd.read_csv(loc_grid)
```

```python
sediments.to_csv(dest_sed_grid)
```

```python
for index, sediment in sediments.iterrows():
    for index2, grid in grids.iterrows():
        if(sediment.LONGITUDE > grid.left and sediment.LONGITUDE < grid.right and sediment.LATITUDE > grid.bottom and sediment.LATITUDE < grid.top):
            sediments.loc[index,'grid_id'] = grid.grid_id 
```
