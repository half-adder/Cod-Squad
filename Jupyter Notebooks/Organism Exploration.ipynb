```python
import pandas as pd

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams['figure.figsize'] = (16,8)

pd.set_option('max_columns', 1000)
```

```python
def fix_lat_lon(series):
    return series.astype(str).str.replace('.', '').str[:-1].astype(int)/100000
```

```python
organism = pd.read_excel('../data/organism_values_201880.xlsx')
# catch_all = pd.read_csv('../data/catch_data/combined_cod_catch_data.csv')
cruises = pd.read_excel('../data/IBS2_First_2_Years/Cruises.xlsx')
species_weights = pd.read_excel('../data/IBS2_First_2_Years/Species_weights.xlsx')
```

```python
tow = pd.read_excel('../data/IBS2_First_2_Years/Tow_Info.xlsx')
tow['START_LAT'] = fix_lat_lon(tow['START_LAT'])
tow['START_LON'] = fix_lat_lon(tow['START_LON'])
tow['END_LAT'] = fix_lat_lon(tow['END_LAT'])
tow['END_LON'] = fix_lat_lon(tow['END_LON'])
```
